if we want to create a new array or DataFrame filled with zeros to store aggregated results, perform calculations, or fill missing values. In such cases, we can use the existing NumPy and Pandas functions to achieve this

```python
# other way of checking aggregated results 

df_setosa = df.loc[df['Species']=='Iris-setosa']
df_verginica = df.loc[df['Species']=='Iris-virginica']
df_versicolor = df.loc[df['Species']=='Iris-versicolor']
```

```python
plt.plot(df_setosa['SepalLengthCm'],np.zeros_like(df_setosa['SepalLengthCm']),'or')
plt.plot(df_verginica['SepalLengthCm'],np.zeros_like(df_verginica['SepalLengthCm']),'oy')
plt.plot(df_versicolor['SepalLengthCm'],np.zeros_like(df_versicolor['SepalLengthCm']),'ob')
plt.xlabel('Sepal Length')
plt.show()
```
