## Univariate , Bivariate and Multivariate Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df.shape
```

### Univariate Analysis

```python
df_setosa = df.loc[df['Species']=='Iris-setosa']
```

```python
df_setosa.head()
```

```python
df_verginica = df.loc[df['Species']=='Iris-virginica']
```

```python
df_verginica.head()
```

```python
df_versicolor = df.loc[df['Species']=='Iris-versicolor']
```

```python
df_versicolor.head()
```

```python
df
```

```python
plt.plot(df_setosa['SepalLengthCm'],np.zeros_like(df_setosa['SepalLengthCm']))
plt.plot(df_verginica['SepalLengthCm'],np.zeros_like(df_verginica['SepalLengthCm']))
plt.plot(df_versicolor['SepalLengthCm'],np.zeros_like(df_versicolor['SepalLengthCm']))
plt.xlabel('Sepal Length')
plt.show()
```

```python
plt.plot(df_setosa['SepalLengthCm'],np.zeros_like(df_setosa['SepalLengthCm']),'o')
plt.plot(df_verginica['SepalLengthCm'],np.zeros_like(df_verginica['SepalLengthCm']),'o')
plt.plot(df_versicolor['SepalLengthCm'],np.zeros_like(df_versicolor['SepalLengthCm']),'o')
plt.xlabel('Sepal Length')
plt.show()
```

### Bivariate Analysis
