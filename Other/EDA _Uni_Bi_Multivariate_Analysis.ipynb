# Bivariate & Multivariate Analysis

The analysis depened on data type (numerical or catagorical):

For visualization below tricks:

##### 1. Numerical - Numerical :
    a. scatterplot
    b. Lineplot
##### 2. Numerical - Categorical 
    a. barplot
    b. boxplot
    c. distplot
##### 3. Catagorical - Catagorical :
    a. HeatMap
    b. clustermap
##### 4. Numerical - Numerical - Numerical - Categorical -  Catagorical - Catagorical 
    a. Pairplot

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

# Numerical - Numerical

### 1. Scatterplot ()

```python
sns.scatterplot(x=df['Age'],y=df['Fare'],hue=df['Survived'])
```

### 2. Lineplot ()

```python
sns.lineplot(x=df['Age'],y=df['Fare'])
```

# Numerical - Categorical

### 1. Bar Plot ()

```python
df.head()
```

```python
sns.barplot(x=df['Pclass'],y=df['Age'])
```

```python
sns.barplot(x=df['Pclass'],y=df['Fare'])
```

```python
sns.barplot(x=df['Pclass'],y=df['Fare'],hue=df['Sex'])
```

### 2. Box Plot ()

```python
sns.boxplot(x=df['Sex'],y=df['Age'],hue=df['Survived'])
```

### 3. Distplot ()

```python
sns.distplot(df['Age'])
```

```python
sns.distplot(df[df['Survived']==0]['Age'],hist=False)
sns.distplot(df[df['Survived']==1]['Age'],hist=False)
```

# Categorical - Categorical

### 1. HeatMap ()

```python
df.head()
```

```python
sns.heatmap(pd.crosstab(df['Pclass'],df['Survived']))
```

### 2. ClusterMap ()

```python
pd.crosstab(df['SibSp'],df['Survived'])
```

```python
sns.clustermap(pd.crosstab(df['Parch'],df['Survived']))
```
