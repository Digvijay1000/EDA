###### univariate analysis

```python
# MEDV - Median Value of owner-Ocuupied Homes (Target Variable)

sns.distplot(df['MEDV'], label='Skewness :%.2f'%(df['MEDV'].skew()))
plt.legend(loc='best')
plt.title('Median Value of Owner-occumpied Homes Distribtion')
```

distribution is close to normal distribution with a slight skewness

```python
log_data = np.log(df['MEDV'])
valid_data = log_data[log_data>0]

std = (valid_data - np.mean(valid_data))/ np.std(valid_data)


sns.distplot(std ,label='Skewness :%.2f'%(std.skew()) )
plt.legend(loc='best')
plt.title('Median Value of Owner-occumpied Homes Distribtion')
```

```python
# CRIM (Per Caipta Crime Rate by Town) Column

sns.distplot(df['CRIM'],label='Skewness :%.2f'%(df['CRIM'].skew()) )
plt.legend(loc='best')
plt.title('Per Capita Crime Rate by Town')

```

 there is a high degree of skewness in the distribution 

```python
# adding constant 
# data_with_const = 1e-10 + df['CRIM'] 

# log transformation
# log_data=np.log(data_with_const)

# handle nan or infinite values
# valid_data = pd.to_numeric(log_data)
# valid_data = np.nan_to_num(log_data)

# handle negative or zero values
log_data = np.log(df['CRIM'])
valid_data = log_data[log_data>0]

std = (valid_data - np.mean(valid_data))/ np.std(valid_data)


sns.distplot(std ,label='Skewness :%.2f'%(std.skew()) )
plt.legend(loc='best')
plt.title('Per Capita Crime Rate by Town')
```

Distribution is close to normal distribution

```python
# ZN - Proportion of Residential Land Zone

sns.distplot(df['ZN'], label='Skewness :%.2f'%(df['ZN'].skew()))
plt.legend(loc='best')
plt.title('Proportion of Residential Land Zone')
```

###### Bivariate Analysis

```python
# MEDV VS CHAS - Mean value per CHAS(1 if the tract bounds the river; 0 otherwise)

mean_sales_price = df[['CHAS','MEDV']].groupby('CHAS',as_index=False).mean()
mean_sales_price
```

```python
sns.barplot(x=mean_sales_price['CHAS'],y=mean_sales_price['MEDV'])
plt.title('Mean Sales Price for each CHAS')
plt.ylabel('Mean Sales Price')
plt.xlabel('CHAS')
```

```python
# MEDV vs RAD - Mean Price per RAD
mean_price_rad = df[['RAD','MEDV']].groupby('RAD', as_index=False).mean()
mean_price_rad
```

```python
sns.barplot(x=mean_price_rad['RAD'],y=mean_price_rad['MEDV'])
plt.title('Mean Sales Price for each RAD')
plt.xlabel('RAD')
plt.ylabel('Mean Sales Price (MEDV)')
```
